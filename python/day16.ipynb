```python
#string formatting 
def pri(name,age):
    return "Hi {}.you age is{}".format(name,age)
    
print(pri("ajith",23))
```

```python
#string formatting 
def pri(name,age):
    return "Hi {a}.you age is{b}".format(a =name,b =age)
    
print(pri("ajith",23))
```

```python
#string formatting 
def pri(name,age):
    return f"Hi {name}.you age is{age}"
    
print(pri("ajith",23))
```
