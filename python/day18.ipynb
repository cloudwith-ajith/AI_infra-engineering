```python
class Error(Exception):
    pass
class Doberror(Error):
    pass
    
```

```python
age = int(input("Enter the age: "))
try:
    if age >=18:
        print("allowed")
    else:
        raise Doberror
except Doberror:
    print("Age should be above 18")
```

```python
class Car():
    def __init__(self,windows,door,etype):
        self.windows = windows
        self.door    = door
        self.etype   = etype
    def Adas(self):
        if self.etype == "petrol":
            print("yes, it has ADAS function")
        else:
            print("NO ADAS")
        
```

```python
car1 = Car(4,5,"diesel")
```

```python
car1.door
```

```python
car1.etype
```

```python
car1.Adas()
```

```python
class Audi(Car):
    def __init__(self,windows,door,etype,topspeed):
        super().__init__(windows,door,etype)
        self.topspeed = topspeed
    def Top_speed(self):
        print(f"{self.topspeed} is the top speed..")
    
```

```python
audiq7 = Audi(4,5,"petrol",198)
```
