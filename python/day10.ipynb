```python
import matplotlib.pyplot as plt
x =[1,2,3,4,5]
z =[]
for m in x:
    n=m*2
    z.append(n)
print(z)

plt.plot(x,z,'g',)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b--')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'r--')
plt.subplot(2,2,4)
plt.plot(x,y,'g--')
```

```python
import numpy as np
x = np.arange(1,11)
y = x*2
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,6*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
lab = "a","b","c","d"
size = [20,40,7,60]
col =["gold","yellowgreen","lightcoral","lightskyblue"]
explode = (0,0,0,0.1)

plt.pie(size,explode=explode,labels=lab,colors=col,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show
```

```python
import matplotlib.pyplot as plt

marks = [10, 15, 20, 25, 30, 35, 40, 45, 100]

plt.boxplot(marks,patch_artist=True)
plt.title("Student Marks")
plt.show()
```
