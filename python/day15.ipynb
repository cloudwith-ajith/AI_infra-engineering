```python
def evernodd(n):
    if n % 2 == 0:
        return "even"
    else:
        return "odd"
print(evernodd(4))
```

```python
# What  if  the list needs to perform the operation 
l1 =[1,2,3,4,5]
```

```python
#map function which is used to perform the operation to every element in the list 
```

```python
# var = map(function,iteralbe --> like list,tuple...)
r = map(evernodd,l1)
print(r)
```

```python
print(list(r))
```

```python
# we can convert the list element from int to string......
l2 =["1","2","3","4","5"]
r  = map(int,l2)
print(list(r))
```

```python
#double the number
v = map(lambda num : num*2,l1)
print(list(v))
```

```python
#filter function which is used to give the item that satisfies the condition 
#syntax
#filter(function, iterable) ---> function returns true or false. If true, keep the item; else, remove the item 
```

```python
l3 =[1,2,3,4,5]
result = filter(lambda num : num % 2==0,l3)
print(list(result))
```

```python
#list Comprehension
#syntax --->[operation for if]
```

```python
list1=[1,2,3,4,5]
[x*x for x in list1]
```

```python
list1=[1,2,3,4,5]
[x*x for x in list1 if x %2 == 0]
```
