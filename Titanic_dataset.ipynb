```python
def impute_age(cols):
    age =cols['Age']
    pclass = cols['Pclass']
    if pd.isnull(age):
        if pclass == 1:
            return 37
        elif pclass == 2:
            return 29;
        else:
            return 24
    else:
        return age
```

```python
train['Age'] = train[['Age','Pclass']].apply(impute_age,axis=1)
```
