```python
class Silicon:
    def __init__(self,silicon_name,silicon_type,build_phase):
        self.silicon_name = silicon_name
        self.silicon_type = silicon_type
        self.build_phase = build_phase
    def siliconname(self):
        return f"The name of the silicon is {self.silicon_name}"
    def silicontype(self):
        return f"The type is {self.silicon_type}"
    def siliconphase(self):
        return f"Silicon in {self.build_phase}"
```

```python
# how to initiate the class 
s1 = Silicon("GNRD", "server-based", "End-phase")
```

```python
s1.build_phase
```

```python
s1.siliconname()
```

```python
s1.siliconphase()
```

```python
s1.silicontype()
```

```python
try:
    a = b
except NameError as x:
    print(" b is not defined")
except Exception as ex:
    print(ex)
```

```python
try:
    a = 1
    b= '2'
    c = a+b
except NameError as x:
    print(" b is not defined")
except Exception as ex:
    print(ex)
```

```python
try:
    a = 1
    b= '2'
    c = a+b
except NameError as x:
    print(" b is not defined")
except TypeError:
    print("only same datatype can be added")
except Exception as ex:
    print(ex)
```
